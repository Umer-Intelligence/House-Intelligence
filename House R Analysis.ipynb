# Loading Libraries

```r
library(dplyr)
library(ggplot2)
```

# Viewing dataset

```r
House <- read.csv('C:/Desktop/House/Houses.csv')
House
```

# Analyzing dataset

```r
#To find the relationship between number of floors and bedrooms
```

```r
House %>% select('bedrooms','floors')
```

```r
#To find the house and the city having the highest price.
```

```r
House %>% select('city','price') %>% filter(price==max(price))
```

```r
#To find the prices of houses over the years
```

```r
House %>% group_by(House$yr_built) %>% summarise(price=mean(price))
```

```r
#To find the number of houses having respective condition
```

```r
House %>% group_by(House$condition) %>% summarise(n())
```

```r
#To find the number of houses having waterfronts
```

```r
House %>% group_by(waterfront) %>% summarise(n())
```

```r
#To create a new column showing price per sqfeet of the house
```

```r
House['Price per sqfeet'] = House$price/House$sqft_lot
House['Price per sqfeet']
```

```r
#To arrange the dataset in the increasing order of price
```

```r
House %>% arrange(price) 
```

```r
#To find the number of houses having respective floors
```

```r
House %>% group_by(House$floors) %>% summarise(n())
```

# Visualizing Dataset

```r
#To find the relationship between price and year_built
```

```r
ggplot(data=House,aes(x=yr_built,y=price)) + geom_bar(stat='identity',color='yellow')
```

```r
#To find the freqency of number of floors
```

```r
hist(x=House$floors,col='red')
```

```r
#To find the frquency of number of bedrooms
```

```r
hist(x=House$bedrooms,col='green')
```

```r
plot(x=House$floors,y=House$bedrooms,col='blue',cex=2)
```
